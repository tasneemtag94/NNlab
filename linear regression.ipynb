```python
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
num_puntos=100
con_puntos=[]
for i in range (num_puntos):
    x=np.random.normal(0.0,0.55)
    y=x*.1+.0+np.random.normal(0.0,0.03)
    con_puntos.append([x,y])
x_data=[v[0] for v in con_puntos]
y_data=[v[1] for v in con_puntos]
plt.plot(x_data,y_data,'r+',label='original data')
plt.legend
plt.show()
```
